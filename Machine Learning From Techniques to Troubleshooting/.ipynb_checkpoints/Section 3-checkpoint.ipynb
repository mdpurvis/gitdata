# Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = ird.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=135)
```

```python
X_train.shape
```

To apply logistic regression use the code below.
What happens now is that it takes each data point at x and looks at its correct label in y.  It then fits the model so it minimizes the error of the predictive model, it gives yo ua bunch of helpful paramaters of how it isn't instantiate it logistic regression and what we can test it is with the second cord below this cell.

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_train[0].reshape(1,-1))
```

```python
y_train[0]
```

How do we evaluate these.  The easiest method is to use the below code: This will help to show the scores of each value.

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
```
